## Regularization

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=0)
```

## L1 Regression

```python
lasso_model=Lasso()
```

```python
lasso_model.fit(x_train,y_train)
```

```python
y_predict_lasso=lasso_model.predict(x_test)
```

```python
mean_squared_error(y_test,y_predict_lasso)
```

```python
lassoScore=lasso_model.score(x_test,y_test)
```

```python
lasso_model.score(xtrain,ytrain)
```

## L2 Regression

```python
ridge_model=Ridge()
```

```python
ridge_model.fit(x_train,y_train)
```

```python
y_predict_ridge=ridge_model.predict(x_test)
```

```python
mean_squared_error(y_test,y_predict_ridge)
```

```python
ridgeScore = ridge_model.score(x_test, y_test)
```

```python
ridge_model.score(x_train,y_train)
```

The answer of ridge,lasso and linear regression are same. Ridge,Lasso is useful in big datasets which will be high dimensional and for dimension reduction we use Ridge,Lasso. But here all te features are imp. So,no feature will be reduced and the answer will be same for ridge,lasso and linear regression

```python
models=['Lasso','Ridge']
scores=[lassoScore,ridgeScore]
```

```python
for i,score in enumerate(scores):
    plt.text(i, score+.05, f"{score:.3f}",ha='center')

plt.bar(models,scores)
plt.ylabel("RÂ² Score")
plt.title("Regression Model Evaluation")
plt.ylim(0,1)
plt.show()
```
