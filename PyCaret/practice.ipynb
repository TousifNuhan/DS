## Random Over Sampler

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros=RandomOverSampler(random_state=0)
```

```python
X= df.drop('Patient_Status',axis=1)
```

```python
y= df[['Patient_Status']]
```

```python
x_resampled,y_resampled = ros.fit_resample(X,y)
```

```python
x_resampled.head()
```

```python
x_resampled.shape
```

```python
y_resampled.shape
```

```python
df2=pd.concat([x_resampled,y_resampled],axis=1)
```

```python
df2.head()
```

```python
df2.Patient_Status.value_counts()
```

```python
model_setup2=setup(
    df2,
    target='Patient_Status',
    fold=5,
    log_experiment=False,
    normalize=True,
    transformation=True,
    remove_multicollinearity=True,
    numeric_imputation='median',
    train_size=0.70,
    memory=None
)
```

```python
best= compare_models()
```

```python
et = create_model('et')
```

```python
evaluate_model(best)
```
