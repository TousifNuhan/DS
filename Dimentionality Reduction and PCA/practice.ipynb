## Wrapper Mthod - Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
rfe= RFE(estimator=classifier,n_features_to_select=10)
```

```python
rfe.fit(X,y)
```

```python
feature_ranks = rfe.ranking_
selected_features = rfe.support_
```

```python
feature_ranks
```

```python
selected_features
```

```python
for feature , rank in zip(X.columns,feature_ranks):
    print(f"Feature:{feature} , Rank:{rank}")
```

## Embedded Method 

L1 Regularization (Lasso)

```python
classifier = LogisticRegression(penalty='l1',solver='liblinear')
```

```python
classifier.fit(X,y)
```

```python
coefficients = classifier.coef_
```

```python
for feature , coefficient in zip(X.columns,coefficients[0]):
    print(f"Feature:{feature} , Coefficient:{coefficient}")
```

## PCA

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mnx= MinMaxScaler()
```

```python
scaled_x = mnx.fit_transform(X)
```

```python
scaled_x
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
x_pca = pca.fit_transform(scaled_x)
```

```python
x_pca
```

```python
features = pd.DataFrame(x_pca,columns=['pca1','pca2','pca3'])
```

```python
features
```

```python
import plotly.express as px
```

```python
px.scatter_3d(features,x='pca1',y='pca2',z='pca3',color='pca1')
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(features,y,test_size=0.30, random_state=12)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```
