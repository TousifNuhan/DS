```python
import pandas as pd
df = pd.read_csv("Dhaka Rent.csv", sep=';')
df.head()
```

# Do experiment!

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
x=df[['area']]
```

```python
y=df['rent']
```

```python
model=LinearRegression().fit(x,y)
```

```python
model
```

```python
mod_coef=model.coef_
```

```python
mod_intercept=model.intercept_
```

```python
df['Predicted']=model.predict(x)
```

```python
df
```

```python
df2=df.copy()
```

```python
model.score(x,y)
```

```python
model.predict([[4000]])
```

```python
df2
```

```python
plt.figure(figsize=(10,7))

m=mod_coef[0]
c=mod_intercept
equ=f"y={float(m):.2f}x + {float(c):.2f}"

plt.text(min(df['area']),max(df['rent'])-10000,equ)

plt.plot(df[['area']],df[['Predicted']],color="black")

plt.scatter(df['area'],df['Predicted'],marker="*",color="red",label="Predicted")

plt.scatter(df['area'],
            df['rent'],
            c=df['rent'],
            cmap='coolwarm',
            s=60,
            alpha=.90,
            edgecolors='g',
            label="rent"
            )

plt.xlabel("Area in Dhaka",fontsize=10)
plt.ylabel("Rent in Dhaka",fontsize=10)
plt.title("Area & Rent in Dhaka",fontsize=15,weight='semibold')

plt.legend()
plt.show()
```
