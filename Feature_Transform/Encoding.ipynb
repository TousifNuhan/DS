## One-Hot Encoding with Label Binarizer

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
```

```python
area_Encoded=lb.fit_transform(df4['Area'])
```

```python
area_encoded_df=pd.DataFrame(area_Encoded,columns=lb.classes_)
```

```python
area_encoded_df.head()
```

```python
area_encoded_df.drop(columns=lb.classes_[0],inplace=True)
```

```python
area_encoded_df.head()
```

```python
df4_encoded=pd.concat([df4.drop('Area',axis=1),area_encoded_df],axis=1)
```

```python
df4_encoded.head()
```

## Mean Encoding

```python
mean_encoded_area = df6.groupby('Area')['Profit'].mean().to_dict()
```

```python
mean_encoded_area
```

```python
df6['Area']= df6['Area'].map(mean_encoded_area)
```
