# Techniques to perform Feature Transformation:

1. Normalization <br>
2. Standardization <br>
3. Log Transformation <br>
4. Robust Scaler <br>
5. Max Absolute Scaler


1. Normalization <br>



```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mns= MinMaxScaler()
```

```python
df['cost_norm']=mns.fit_transform(df[['cost']])
```

```python
df
```

2. Standardization

```python
mean_cost= df.cost.mean()
std_cost= df.cost.std()
```

```python
df['cost_std_manual']=(df['cost'] - mean_cost)/std_cost
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdS= StandardScaler()
```

```python
df['cost_std_SKLEARN']=stdS.fit_transform(df[['cost']])
```

```python
df
```

```python
std_ddof0 = df.cost.std(ddof=0)
```

```python
std_ddof0
```

```python
std_ddof1= df.cost.std(ddof=1)
```

```python
std_ddof1
```

```python
df['cost_std_manual_1']=(df['cost'] - mean_cost)/std_ddof1
```

```python
df
```

```python
df['cost_std_SKLearn_1']= (df['cost'] - mean_cost)/std_ddof0
```

```python
df
```

3. Log Transformation

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
```

```python
fc_t = FunctionTransformer(np.log1p)
```

```python
df['fc_t'] = fc_t.fit_transform(df[['cost']])
```

```python
df
```

```python
## another way

df['fc_t_np'] = np.log1p(df[['cost']])
```

```python
df
```

## Robust Scalar

```python
Q1_25= np.percentile(df['cost'],25)
```

```python
Q2_50=np.percentile(df['cost'],50)
```

```python
Q3_75 = np.percentile(df['cost'],75)
```

```python
median_cost= df.cost.median()
```

```python
IQR = Q3_75 - Q1_25
```

```python
df['RobustScalar']= (df['cost'] - median_cost) / IQR
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Standard_Scaler=StandardScaler()
```

```python
df['RobustScalar_sklearn'] = Standard_Scaler.fit_transform(df[['cost']])
```

```python
df
```

# Max Abs Scaler

```python
df['MAError_Manual']= df['cost'] / max(df['cost'])
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
MAErroroBJ=MaxAbsScaler()
```

```python
df['MAError_SKLearn']= MAErroroBJ.fit_transform(df[['cost']])
```

```python
df
```
